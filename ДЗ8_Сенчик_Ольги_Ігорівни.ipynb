Ти розглядаєш можливість інвестувати в стартап-компанію, яка розробляє новий продукт на ринку. Маєш два варіанти вкладень: варіант А та варіант В.

Наведено ймовірності (p ) та грошові потоки (x, тис дол) за умови різних сценаріїв:

Яка компанія, з точки очікуваної прибутковості (математичне очікування) є цікавішою при таких даних?

telegram-cloud-photo-size-2-5400294292771165261-y.jpg

```python
probabilities_A = [0.05, 0.25, 0.3, 0.4]
probabilities_B = [0.2, 0.25, 0.25, 0.3]

flows_A = [0, 25, 40, 80]

flows_B = [10, 30, 60, 70]

expected_value_A = sum(p * x for p, x in zip(probabilities_A, flows_A))

expected_value_B = sum(p * x for p, x in zip(probabilities_B, flows_B))

print("Математичне очікування для варіанту A:", expected_value_A)
print("Математичне очікування для варіанту B:", expected_value_B)

print("Висновок: ")
if expected_value_A > expected_value_B:
    print("З точки зору математичного очікування, варіант A є кращим.")
elif expected_value_A < expected_value_B:
    print("З точки зору математичного очікування, варіант B є кращим.")
else:
    print("Варіанти A та B мають однакове математичне очікування.")
```

# **Завдання 4**

Тебе зацікавили можливості підтосовування результатів: досліди, чи це можливо. Згенеруй нормально-розподілену величину і зміни декілька показників так, щоб ряд перестав проходити тест на нормальний розподіл.

```python
mean = 5
std_dev = 2
sample_size = 100

X = np.random.normal(mean, std_dev, sample_size)

plt.hist(X, bins=10, color='pink', edgecolor='black', alpha=0.7)
plt.xlabel('Значення')
plt.ylabel('Частота (йомовірність)')
plt.title('Діаграма розподілу (початковий)')
plt.show()

k2_initial, pvalue_initial = stats.normaltest(X)

num_changes = 10
indices_to_change = np.random.choice(sample_size, num_changes, replace=False)
X[indices_to_change] = np.random.normal(mean + 10, std_dev * 5, num_changes)

plt.hist(X, bins=10, color='pink', edgecolor='black', alpha=0.7)
plt.xlabel('Значення')
plt.ylabel('Частота (йомовірність)')
plt.title('Діаграма розподілу (змінений)')
plt.show()

k2_changed, pvalue_changed = stats.normaltest(X)
if pvalue_initial > 0.05 and pvalue_changed <= 0.05:
    print("Початковий розподіл схожий на нормальний, але змінений не проходить перевірку на нормальність")
else:
    print("Умова не виконана")
```
