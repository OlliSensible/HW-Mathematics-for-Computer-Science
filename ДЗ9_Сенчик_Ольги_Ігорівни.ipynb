Ти придбав компанію, в якій колись працював. Компанія ледь-ледь зводить кінці з кінцями, хоча процеси в ній вибудувано. Дослідивши їх, ти зрозумів, що ті сервіси, які компанія пропонує, не є оптимальними. Наразі компанія виконує три різних послуги, дохід від яких не однаковий, і через відсутність концентрації на найприбутковішому продукті компанія недоотримує прибуток.

Ти вже оптимізував роботу компанії завдяки тому, що обрав той набір сервісів, який приносить найбільший прибуток, зважаючи на можливості роботи команди. Наразі твоїм завданням є формулювання задачі оптимізації процесу розробки в компанії за нижченаведених умов.

В процесі задіяні дизайнери, розробники та тестувальники. Дизайнерів в компанії двоє, фонд праці кожного 160 годин роботи на місяць. Програмістів також двоє, проте їх загальний лише 192 години. Тестувальник один, але працює понадномово - 180 годин.

Компанія може створювати сайти, інтернет-магазини та інтеграції із ERP. Вартість послуги розробки та час роботи спеціалістів за кожним видом робіт наступні:

```python
from scipy.optimize import linprog

c = [-2, -9, -6]

A = [
    [12, 12, 12],
    [6, 24, 18],
    [2, 18, 12],
]

b = [320, 192, 180]

x0_bounds = (0, None)
x1_bounds = (0, None)
x2_bounds = (0, None)

result = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds, x2_bounds], method='highs')

print("Оптимальний розподіл робіт:")
print(f"Сайти: {result.x[0]} робіт")
print(f"Інтернет-магазин: {result.x[1]} робіт")
print(f"Інтеграція з ERP: {result.x[2]} робіт")
print("\nСумарний місячний дохід компанії: ${:.2f} тис.".format(-result.fun))
```
