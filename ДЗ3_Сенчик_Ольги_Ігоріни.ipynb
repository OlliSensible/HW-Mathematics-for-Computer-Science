# **Task 1**

If we consider the billiard table as a plane, the velocity of the billiard ball before hitting the cushion was (3,2). The impact on the cushion imparted a velocity to the ball that can be described by the vector (0,-4). What is the velocity of the ball after the impact? Formulate the calculations for this task graphically and using the NumPy library.

Calculate the angle at which the ball struck, too.

```python
import numpy as np
import matplotlib.pyplot as plt

# Початкова швидкість кулі до удару в борт
# Значення "3" - швидкість по горизонталі.
# Значення "2" - швидкість по вертикалі.
v_before = np.array([3, 2])

# Зміна швидкості після удару в борт
# Значення "0" - швидкість по горизонталі тобто її відсутність.
# Значення "-4" - швидкість по вертикалі і вказує на зменшення швидкості вниз.
v_strike = np.array([0, -4])

v_after = v_before + v_strike

angle_rad = np.arccos(np.dot(v_before, v_after) / (np.linalg.norm(v_before) * np.linalg.norm(v_after)))
angle_deg = np.degrees(angle_rad)

origin = np.zeros(2)

plt.quiver(*origin, *v_before, scale=1, scale_units='xy', angles='xy', color='#FF9496', label='До удару')
plt.quiver(*origin, *v_strike, scale=1, scale_units='xy', angles='xy', color='#f73339', label='Удар')
plt.quiver(*origin, *v_after, scale=1, scale_units='xy', angles='xy', color='#2596be', label='Після удару')

plt.xlim(-1, 5)
plt.ylim(-5, 3)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=1)
plt.legend()
plt.title(f"Кут удару: {angle_deg:.2f} градусів")
plt.show()

```

# **Task** 3

Goods in the industry are produced with a velocity described by the vector (1,5,2,6,7). Due to negative force majeure events, the changes for the current year can be characterized as (0,-1,0,-0,-1). However, the industry maintains a growth rate of 20% per year. In how much time will the industry completely overcome the consequences of force majeure circumstances?

```python
# Задаємо вектор продуктивності до форс-мажору
a = np.array([1, 5, 2, 6, 7])
# Задаємо вектор поточного року з форс-мажором
b = np.array([0, -1, 0, 0, -1])

all_positive_a = all(a > 0)
da = 0.2 * a

change = np.zeros_like(a, dtype=np.float64)

while not all(change - (- b) > 0):
    change += da

print(f"Вектор швидкості (a): {a}")
print(f"Вектор змін (b): {b}")
print(f"Вектор темпу (da): {da}")
print(f"Кількість років для подолання форс-мажорів: {len(change)} років")
```
