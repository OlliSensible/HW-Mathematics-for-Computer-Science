Обрахуй значення цінності, яку приносить компанії розробник за 8 годин робочого дня.

```python
from sympy import symbols, diff, solve
import numpy as np
import matplotlib.pyplot as plt

x = symbols('x')
efficiency_function = -3*x**2 + 30*x

derivative_efficiency = diff(efficiency_function, x)
print(f"Похідна функції ефективності:")
derivative_efficiency
```

```python
def efficiency(x):
    return -3 * x**2 + 30 * x

def derivative_efficiency(x):
    return -6 * x + 30

x_values = np.linspace(0, 10, 100)

y_values = efficiency(x_values)
dy_values = derivative_efficiency(x_values)

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values, label='Функція ефективності')
plt.plot(x_values, dy_values, label='Похідна функції')

plt.title('Графік функції та похідної')
plt.xlabel('Години роботи на день')
plt.ylabel('Ефективність / Похідна ефективності')
plt.legend()

plt.grid(True)
plt.show()
```

```python
def efficiency(x):
    return -3*x**2 + 30*x

x = symbols('x')

y = -3*x**2 + 30*x

derivative_y = diff(y, x)

x_max = solve(derivative_y, x)[0]

y_max = efficiency(x_max)

print(f"Точка максимуму: ({x_max}, {y_max})")
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values, label='Функція ефективності')
plt.scatter(x_max, y_max, color='red', label='Точка максимуму')

plt.title('Графік функції та точки максимуму')
plt.xlabel('Години роботи на день')
plt.ylabel('Ефективність')
plt.legend()

plt.grid(True)
plt.show()
```

```python
x_8_hours = 8
y_8_hours = efficiency(x_8_hours)

print(f"Ефективність за 8 годин робочого дня: {y_8_hours}")
```

# **Завдання 2**

Ти сформував політику взаємодії між співробітниками, тімлідами та проджект менеджерами і отримав такий результат, що функція ефективності роботи співробітників змінилася. Тепер вона має вигляд

image.png

Оціни ефективність після зміни схеми роботи співробітників: на скільки більше став приносити доходу компанії один співробітник за 8 годин роботи на день?



```python
x = symbols('x')
efficiency_function = -3 * (x - 2)**2 + 30 * x

derivative_efficiency = diff(efficiency_function, x)
print(f"Похідна функції ефективності:")
derivative_efficiency
```

```python
efficiency_np = np.vectorize(lambda x_val: -3 * (x_val - 2)**2 + 30 * x_val)
derivative_efficiency_np = np.vectorize(lambda x_val: derivative_efficiency.subs(x, x_val))

x_values = np.linspace(0, 10, 100)

y_values = efficiency_np(x_values)
dy_values = derivative_efficiency_np(x_values)

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values, label='Функція ефективності')
plt.plot(x_values, dy_values, label='Похідна функції')

plt.title('Графік функції та похідної')
plt.xlabel('Години роботи на день')
plt.ylabel('Ефективність / Похідна ефективності')
plt.legend()

plt.grid(True)
plt.show()
```

```python
critical_points = solve(derivative_efficiency, x)
max_point = (critical_points[0], efficiency_function.subs(x, critical_points[0]))

print(f"Точка максимуму: {max_point}")
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values, label='Функція ефективності')
plt.scatter(max_point[0], max_point[1], color='red', label='Точка максимуму')

plt.title('Графік функції та точки максимуму')
plt.xlabel('Години роботи на день')
plt.ylabel('Ефективність')
plt.legend()

plt.grid(True)
plt.show()
```

```python
x_8_hours = 8
y_8_hours = efficiency_function.subs(x, x_8_hours)

print(f"Ефективність за 8 годин робочого дня: {y_8_hours}")
```

```python
previous_efficiency = efficiency(8)

increase_in_efficiency = y_8_hours - previous_efficiency

print(f"Зростання ефективності: {increase_in_efficiency}")
```
