# **Task 1 Matrix Solutions**

Solve a system of linear equations using the matrix method with the help of the inverse matrix:

image.png
image.png

```python
import numpy as np

# Step 1: Define the system of equations in matrix form using NumPy
A = np.array([[-1, 1, 2],
              [0, -1, -3],
              [4, -3, 2]])

B = np.array([1, -4, 7])

# Step 2: Find and print the determinant of the matrix
det_A = np.linalg.det(A)
print(f"Determinant of matrix A: {det_A}")

# Check if the matrix is non-singular (determinant is not zero)
if det_A != 0:
    # Step 3: Find and print the inverse matrix
    A_inv = np.linalg.inv(A)
    print("Inverse matrix of A:")
    print(A_inv)

    # Step 4: Find and print the solution matrix
    X = np.dot(A_inv, B)
    print("Solution of the system of equations:")
    print(X)
else:
    print("The system of equations does not have a unique solution since the determinant of matrix A is zero.")

```

# **Task 3 Solving a Problem with a Textual Condition**

Three types of specialists worked on the project - designers, developers, and testers. We know the number of hours at each stage of development and the amount of money allocated for each stage as the cost.

The manager divided the entire development into three stages: design creation, software development, and final testing.

On the first stage, the designer worked for 20 hours, and the developers worked for 3 hours - the estimated cost for this stage is 800 dollars. On the second stage, designers worked for 8 hours, developers for 28, and testers for 4. This stage was estimated at 900 dollars. On the final stage, designers worked for 4 hours, developers for 12, and testers for 32. The labor resources for this stage were estimated at 600 dollars.

Find the hourly rate for each specialist by solving a system of linear equations based on the data for each development stage.

```python
# Step 1: Define the system of equations in matrix form using NumPy
A = np.array([[20, 3, 0],
              [8, 28, 4],
              [4, 12, 32]])

B = np.array([800, 900, 600])

# Step 2: Find and print the determinant of the matrix
det_A = np.linalg.det(A)
print(f"Determinant of matrix A: {det_A}")

if det_A != 0:
    # Step 3: Find and print the inverse matrix
    A_inv = np.linalg.inv(A)
    print("Inverse matrix of A:")
    print(A_inv)

    # Step 4: Find and print the solution matrix (cost per hour for each specialist)
    X = np.dot(A_inv, B)
    print("Cost per hour for each specialist:")
    print(X)
else:
    print("The system of equations does not have a unique solution since the determinant of matrix A is zero.")
```
