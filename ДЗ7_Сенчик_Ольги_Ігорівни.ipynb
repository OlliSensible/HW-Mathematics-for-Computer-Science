6. Підкидають три гральні кубики. Що ймовірніше: отримати в сумі очок, що випали, 11 або 12?

```python
from itertools import product

dice_values = [1, 2, 3, 4, 5, 6]

all_combinations = list(product(dice_values, repeat=3))

sum_11_combinations = [comb for comb in all_combinations if sum(comb) == 11]
sum_12_combinations = [comb for comb in all_combinations if sum(comb) == 12]

probability_sum_11 = len(sum_11_combinations) / len(all_combinations)
probability_sum_12 = len(sum_12_combinations) / len(all_combinations)

result = " "
if probability_sum_12 > probability_sum_11:
    result = "Ймовірніше отримати 12"
elif probability_sum_12 == probability_sum_11:
    result = "Ймовірніcть отримати 11 і 12 однакова"
else:
    result = "Ймовірніше отримати 11"

print(result)
```

7. На десятьох однакових картках написані числа від 0 до 9. Визнач ймовірність того, що навмання утворене за допомогою даних карток

а) двозначне число ділиться на 18;

б) тризначне число ділиться на 36.

```python
res = ""
count = 0
for i in range(18, 100):
    if i % 18 == 0:
        res += str(i) + " "
        count += 1

print(res)

variant = 100-10

result = count / variant

print("Ймовірність утворення двозначного числа, яке ділиться на 18, становить:", result)
```

```python
res = ""
count = 0
for i in range(100, 1000):
    if i % 36 == 0:
        res += str(i) + " "
        count += 1

print(res)

variant = 1000 - 100

result = count / variant

print("Ймовірність утворення двозначного числа, яке ділиться на 36, становить:", result)
```

```python
from itertools import permutations

def is_divisible_by_18(number_18):
    return number_18 % 18 == 0

def calculate_probability():
    cards = list(range(10))
    permutations_count = 0
    divisible_by_18_count = 0

    for perm in permutations(cards, 2):
        two_digit_number = perm[0] * 10 + perm[1]
        permutations_count += 1
        if is_divisible_by_18(two_digit_number):
            divisible_by_18_count += 1

    probability = (divisible_by_18_count / permutations_count) * 100
    return probability

result = calculate_probability()
print(f"Ймовірність утворення двозначного числа, яке ділиться на 18, становить: {result:.2f}%")
```

```python
from itertools import permutations

def is_divisible_by_36(number_3_digits):
    return number_3_digits % 36 == 0

def calculate_probability():
    cards = list(range(10))
    permutations_count = 0
    divisible_by_36_count = 0

    for perm in permutations(cards, 3):
        three_digit_number = perm[0] * 100 + perm[1] * 10 + perm[2]
        permutations_count += 1
        if is_divisible_by_36(three_digit_number):
            divisible_by_36_count += 1

    probability = (divisible_by_36_count / permutations_count) * 100
    return probability

result = calculate_probability()
print(f"Ймовірність утворення тризначного числа, яке ділиться на 36, становить: {result:.2f}%")
```

На площині накреслені дві концентричні окружності, радіуси яких 5 і 10 см відповідно. Знайди ймовірність того, що точка, кинута навмання у велике коло, потрапить також у кільце, утворене побудованими окружностями.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Створення графа
G = nx.DiGraph()  # Змінено на орієнтований граф

# Додаємо ребра (з'єднання між елементами)
edges = [(1, 3), (1, 5), (3, 4), (2, 3), (2, 5), (4, 6), (5, 6)]
G.add_edges_from(edges)

# Розміщення вершин графа
pos = {1: (0, 1), 2: (0, 0), 3: (1, 1), 4: (1, 0), 5: (2, 1), 6: (2, 0)}

# Відображення графа зі стрілочками
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=1000, node_color='skyblue', connectionstyle='arc3,rad=0.1', arrowsize=20)
plt.show()



```
