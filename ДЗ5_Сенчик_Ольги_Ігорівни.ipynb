```python
from sympy import sqrt

expression = sqrt(x**2 + 1)

derivative = diff(expression, x)

derivative
```

```python
result_at_1 = derivative.subs(x, 1)
result_at_fraction = derivative.subs(x, -1/2)

result_at_1.evalf()
```
